```python
import time
import sys
!{sys.executable} -m pip install xlwt

import pynqp2p
from datetime import datetime

from xlwt import Workbook

from grove import Grove_AHT20
from pynq.lib.pmod import PMOD_GROVE_G3

from pynq.lib.arduino import Grove_Light
from pynq.lib.arduino import ARDUINO_GROVE_I2C

from pynq.overlays.base import BaseOverlay
base = BaseOverlay('base.bit')
```

```python
%history -g
```

```python
py_ht = Grove_AHT20(base.PMODB, PMOD_GROVE_G3)
light_sensor = Grove_Light(base.ARDUINO, ARDUINO_GROVE_I2C)

wb = Workbook()

btn0 = base.buttons[0]
led0 = base.leds[0]
led0.off()
sheet1 = wb.add_sheet('Janurary-1')
```

```python
time_slot = 30 # seconds

led0.on()
sheet1.write(0,1,"year/day-1")
sheet1.write(0,2,"hour;5/minute")
sheet1.write(0,3,"humidity")
sheet1.write(0,4,"temperature")
sheet1.write(0,5,"light")

row = 0
column = 1

while btn0.read() == 0:
    try:
        humidity = py_ht.get_humidity()
        temp_c = py_ht.get_temp()
        print('Recorded Humidity: ', float("{0:.2f}".format(humidity)), '%')
        print('Recorded Temperature: ', float("{0:.2f}".format(temp_c)), '°C')
        print('Recorded Light: ', light_sensor.read())
        print(str(datetime.now().time()))
        b = str(datetime.now().year) + " " + str(datetime.now().day)
        sheet1.write(column, 1, b)
        a = str(datetime.now().hour) + " " + str(datetime.now().minute)
        sheet1.write(column, 2, a)
        sheet1.write(column, 3, float("{0:.2f}".format(humidity)))
        sheet1.write(column, 4, float("{0:.2f}".format(temp_c)))
        sheet1.write(column, 5, light_sensor.read())
        wb.save('backup1.xls')
        column+=1
    except Exception as e:
        sheet1.write(column,1,str(e))
        column+=1
        wb.save('backup1.xls')
    for i in range(time_slot): #implemented so the btn0 press can be detected while waiting for 
        time.sleep(1) #time_slot is initialized above
        if btn0.read() == 1:
            break
led0.off()
wb.save('Save1.xls')
```
